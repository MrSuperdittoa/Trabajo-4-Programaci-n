```python
# 02 - preprocesado.ipynb

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el conjunto de datos
file_path = 'dataset_02052023.xlsx'
df = pd.read_excel(file_path)

# Eliminar columnas no numricas o irrelevantes para el anlisis
df.drop(['Timestamp', 'Num'], axis=1, inplace=True, errors='ignore')

# Convertir la columna 'grip_lost' a int si es necesario
if df['grip_lost'].dtype == 'object':
    df['grip_lost'] = df['grip_lost'].astype(int)

# Verificar si hay valores NaN en la variable objetivo
print(df['Robot_ProtectiveStop'].isna().sum())

# Eliminar filas con valores NaN en la variable objetivo
df.dropna(subset=['Robot_ProtectiveStop'], inplace=True)

# Visualizar la correlacin entre las variables
corr_matrix = df.corr()
plt.figure(figsize=(20, 15))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()

# Preprocesado y limpieza de datos
# Eliminar datos faltantes y duplicados
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)

# Verificar tipos de datos y valores NaN
print(df.info())
print(df.isna().sum())



```
