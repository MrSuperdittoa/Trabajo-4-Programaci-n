```python
# 01 - exploracin.ipynb

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el conjunto de datos
file_path = 'dataset_02052023.xlsx'
df = pd.read_excel(file_path)

# Anlisis exploratorio de datos
print(df.head())
print(df.describe())

# Visualizar la distribucin de las clases
sns.countplot(x='Robot_ProtectiveStop', data=df)
plt.title('Distribucin de Clases')
plt.show()

# Visualizar la distribucin de las variables numricas
df.hist(bins=50, figsize=(20, 15))
plt.show()

```
